```python
import sys
import os

# Adiciona o diretório raiz do projeto ao sys.path
sys.path.append(os.path.abspath('..'))
```

```python
import pandas as pd
from data.merges import (
    merge_executor_with_uf, merge_meta_with_uf, merge_finalidade_with_uf,
    merge_documento_with_uf, merge_ordem_pagamento_with_uf, merge_historico_pagamento_with_uf,
    merge_plano_trabalho_with_uf, merge_relatorio_gestao_with_uf
)
from data.conversions import export_to_excel
from api import endpoints
from data.processing import to_dataframe
```

```python
# Baixar e transformar os dados
df_planos = to_dataframe(endpoints.get_plano_acao_especial())
df_executores = to_dataframe(endpoints.get_executor_especial())
df_metas = to_dataframe(endpoints.get_meta_especial())
df_finalidades = to_dataframe(endpoints.get_finalidade_especial())
df_empenhos = to_dataframe(endpoints.get_empenho_especial())
df_documentos = to_dataframe(endpoints.get_documento_habil_especial())
df_ordens = to_dataframe(endpoints.get_ordem_pagamento_ordem_bancaria_especial())
df_historicos = to_dataframe(endpoints.get_historico_pagamento_especial())
df_planos_trabalho = to_dataframe(endpoints.get_plano_trabalho_especial())
df_relatorios = to_dataframe(endpoints.get_relatorio_gestao_especial())
```

```python
# Merges (sem filtro)
df_executores_uf = merge_executor_with_uf(df_executores, df_planos)
df_metas_uf = merge_meta_with_uf(df_metas, df_executores, df_planos)
df_finalidades_uf = merge_finalidade_with_uf(df_finalidades, df_executores, df_planos)
df_documentos_uf = merge_documento_with_uf(df_documentos, df_empenhos, df_planos)
df_ordens_uf = merge_ordem_pagamento_with_uf(df_ordens, df_documentos_uf, df_empenhos, df_planos)
df_historicos_uf = merge_historico_pagamento_with_uf(df_historicos, df_ordens_uf, df_documentos_uf, df_empenhos, df_planos)
df_planos_trabalho_uf = merge_plano_trabalho_with_uf(df_planos_trabalho, df_planos)
df_relatorios_uf = merge_relatorio_gestao_with_uf(df_relatorios, df_planos)
```

```python
# Exportar para Excel
export_to_excel(df_planos, '../excel_without_filter/plano_acao_especial.xlsx')
export_to_excel(df_executores_uf, '../excel_without_filter/executor_especial.xlsx')
export_to_excel(df_metas_uf, '../excel_without_filter/meta_especial.xlsx')
export_to_excel(df_finalidades_uf, '../excel_without_filter/finalidade_especial.xlsx')
export_to_excel(df_empenhos, '../excel_without_filter/empenho_especial.xlsx')
export_to_excel(df_documentos_uf, '../excel_without_filter/documento_habil_especial.xlsx')
export_to_excel(df_ordens_uf, '../excel_without_filter/ordem_pagamento_ordem_bancaria_especial.xlsx')
export_to_excel(df_historicos_uf, '../excel_without_filter/historico_pagamento_especial.xlsx')
export_to_excel(df_planos_trabalho_uf, '../excel_without_filter/plano_trabalho_especial.xlsx')
export_to_excel(df_relatorios_uf, '../excel_without_filter/relatorio_gestao_especial.xlsx')
```
