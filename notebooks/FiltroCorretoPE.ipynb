```python
import requests
import pandas as pd
```

```python

CNPJ_PE = "10571982000125"
url = "https://api.transferegov.gestao.gov.br/transferenciasespeciais/plano_acao_especial"
params = {
    "cnpj_beneficiario_plano_acao": f"eq.{CNPJ_PE}"
}
response = requests.get(url, params=params)
dados = response.json()

# Se a resposta for uma lista de dicts:
df_planos_pe_2020 = pd.DataFrame(dados)
display(df_planos_pe_2020.head(20))
```

```python
ids_planos_pe_2020 = df_planos_pe_2020['id_plano_acao']

print(ids_planos_pe_2020.head(20))
```

```python
# Supondo que ids_planos_pe_2020 seja uma lista de inteiros ou strings
ids_str = ",".join(str(i) for i in ids_planos_pe_2020)
url = "https://api.transferegov.gestao.gov.br/transferenciasespeciais/plano_trabalho_especial"
params = {
    "id_plano_acao": f"in.({ids_str})"
}
response = requests.get(url, params=params)
dados = response.json()

if isinstance(dados, list):
    df_planos_trabalho_pe_2020 = pd.DataFrame(dados)
else:
    print("Resposta inesperada:", dados)
    df_planos_trabalho_pe_2020 = pd.DataFrame()
display(df_planos_trabalho_pe_2020.head(20))
```

```python
df_planos_pe_2020.to_excel("../excel_with_filter/2020_notebooks/planos_acao_especial_PE_2020.xlsx", index=False)
df_planos_trabalho_pe_2020.to_excel("../excel_with_filter/2020_notebooks/planos_trabalho_especial_PE_2020.xlsx", index=False)
```

```python
df_planos_pe_2020.to_csv("../csv_with_filter/planos_acao_especial_PE_2020.xlsx", index=False)
df_planos_trabalho_pe_2020.to_csv("../csv_with_filter/planos_trabalho_especial_PE_2020.xlsx", index=False)
```

```python
# Carregue os arquivos Excel
df_planos = pd.read_excel("../excel_with_filter/2020_notebooks/planos_acao_especial_PE_2020.xlsx")
df_planos_trabalho = pd.read_excel("../excel_with_filter/2020_notebooks/planos_trabalho_especial_PE_2020.xlsx")

# Faça o merge pelo campo em comum (ajuste se o nome da coluna for diferente)
df_merged = df_planos_trabalho.merge(
    df_planos,
    on="id_plano_acao",
    how="left",
    suffixes=('_trabalho', '_plano')
)

# Exporte o resultado
df_merged.to_excel("../excel_with_filter/2020_notebooks/planos_trabalho_completo_PE_2020.xlsx", index=False)
```
