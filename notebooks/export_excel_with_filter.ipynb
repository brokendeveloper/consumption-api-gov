```python
import sys
import os

# Adiciona o diretório raiz do projeto ao sys.path
sys.path.append(os.path.abspath('..'))
```

```python
import pandas as pd
from data.merges import (
    merge_executor_with_uf, merge_meta_with_uf, merge_finalidade_with_uf,
    merge_documento_with_uf, merge_ordem_pagamento_with_uf, merge_historico_pagamento_with_uf,
    merge_plano_trabalho_with_uf, merge_relatorio_gestao_with_uf
)
from data.conversions import export_to_excel
from api import endpoints
from data.processing import to_dataframe
```

```python
# Baixar e transformar os dados
df_planos = to_dataframe(endpoints.get_plano_acao_especial())
df_executores = to_dataframe(endpoints.get_executor_especial())
df_metas = to_dataframe(endpoints.get_meta_especial())
df_finalidades = to_dataframe(endpoints.get_finalidade_especial())
df_empenhos = to_dataframe(endpoints.get_empenho_especial())
df_documentos = to_dataframe(endpoints.get_documento_habil_especial())
df_ordens = to_dataframe(endpoints.get_ordem_pagamento_ordem_bancaria_especial())
df_historicos = to_dataframe(endpoints.get_historico_pagamento_especial())
df_planos_trabalho = to_dataframe(endpoints.get_plano_trabalho_especial())
df_relatorios = to_dataframe(endpoints.get_relatorio_gestao_especial())
```

```python
# Merges
df_executores_pe = merge_executor_with_uf(df_executores, df_planos)
df_metas_pe = merge_meta_with_uf(df_metas, df_executores, df_planos)
df_finalidades_pe = merge_finalidade_with_uf(df_finalidades, df_executores, df_planos)
df_documentos_pe = merge_documento_with_uf(df_documentos, df_empenhos, df_planos)
df_ordens_pe = merge_ordem_pagamento_with_uf(df_ordens, df_documentos_pe, df_empenhos, df_planos)
df_historicos_pe = merge_historico_pagamento_with_uf(df_historicos, df_ordens_pe, df_documentos_pe, df_empenhos, df_planos)
df_planos_trabalho_pe = merge_plano_trabalho_with_uf(df_planos_trabalho, df_planos)
df_relatorios_pe = merge_relatorio_gestao_with_uf(df_relatorios, df_planos)
```

```python
# Filtro PE
df_planos_pe = df_planos[df_planos['uf_beneficiario_plano_acao'] == 'PE']
df_executores_pe = df_executores_pe[df_executores_pe['uf_beneficiario_plano_acao'] == 'PE']
df_metas_pe = df_metas_pe[df_metas_pe['uf_beneficiario_plano_acao'] == 'PE']
df_finalidades_pe = df_finalidades_pe[df_finalidades_pe['uf_beneficiario_plano_acao'] == 'PE']
df_empenhos_pe = df_empenhos[df_empenhos['uf_beneficiario_empenho'] == 'PE']
df_documentos_pe = df_documentos_pe[df_documentos_pe['uf_beneficiario_plano_acao'] == 'PE']
df_ordens_pe = df_ordens_pe[df_ordens_pe['uf_beneficiario_plano_acao'] == 'PE']
df_historicos_pe = df_historicos_pe[df_historicos_pe['uf_beneficiario_plano_acao'] == 'PE']
df_planos_trabalho_pe = df_planos_trabalho_pe[df_planos_trabalho_pe['uf_beneficiario_plano_acao'] == 'PE']
df_relatorios_pe = df_relatorios_pe[df_relatorios_pe['uf_beneficiario_plano_acao'] == 'PE']
```

```python
# Exportar para Excel
export_to_excel(df_planos_pe, '../excel_with_filter/plano_acao_especial_PE.xlsx')
export_to_excel(df_executores_pe, '../excel_with_filter/executor_especial_PE.xlsx')
export_to_excel(df_metas_pe, '../excel_with_filter/meta_especial_PE.xlsx')
export_to_excel(df_finalidades_pe, '../excel_with_filter/finalidade_especial_PE.xlsx')
export_to_excel(df_empenhos_pe, '../excel_with_filter/empenho_especial_PE.xlsx')
export_to_excel(df_documentos_pe, '../excel_with_filter/documento_habil_especial_PE.xlsx')
export_to_excel(df_ordens_pe, '../excel_with_filter/ordem_pagamento_ordem_bancaria_especial_PE.xlsx')
export_to_excel(df_historicos_pe, '../excel_with_filter/historico_pagamento_especial_PE.xlsx')
export_to_excel(df_planos_trabalho_pe, '../excel_with_filter/plano_trabalho_especial_PE.xlsx')
export_to_excel(df_relatorios_pe, '../excel_with_filter/relatorio_gestao_especial_PE.xlsx')
```
